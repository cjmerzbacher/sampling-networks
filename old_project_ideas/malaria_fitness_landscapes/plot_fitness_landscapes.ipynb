```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pyr_data = pd.read_csv('pyr_fitness.csv')
cyc_data = pd.read_csv('cyc_fitness.csv')
pyr_plot = np.array(pyr_data.drop(['Binary', 'Mutation'], axis=1))
cyc_plot = np.array(cyc_data.drop(['Binary', 'Mutation'], axis=1))
list_mutations = list(pyr_data.Mutation)
drug_concentrations_pyr = list(pyr_data.columns[2:])
drug_concentrations_cyc = list(cyc_data.columns[2:])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5), sharey=True)
fig.suptitle('P. falciparum growth rate under drug exposure', fontsize=16)
ax = axs[0]
sns.heatmap(pyr_plot, ax = ax)
ax.set_ylabel('Mutation', fontsize=14)
ax.set_xlabel('Pyrimethamine Concentration', fontsize=14)
ax.set_xticklabels(drug_concentrations_cyc, fontsize=12, rotation=90)
ax.set_yticklabels(list_mutations, fontsize=12, rotation=0)

ax = axs[1]
sns.heatmap(cyc_plot, ax = ax)
ax.set_xticklabels(drug_concentrations_pyr, fontsize=12, rotation=90)
ax.set_yticklabels(list_mutations, fontsize=12, rotation=0)
ax.set_xlabel('Cycloguanil Concentration', fontsize=14)

fig.tight_layout()
```
