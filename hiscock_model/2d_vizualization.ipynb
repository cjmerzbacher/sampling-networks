```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('2d_sample_run.csv')
data
```

```python
initial = data.loc[data.conv == 'initial']
```

```python
classes = list(data['class'])
total_scores = list(data['score'])
print('Total domains of attraction found: ', len(set(classes)))
print('Locations of stable steady states: ', list(set(classes)))

fig, axs = plt.subplots(1, 2, figsize=(10,4))
ax = axs[0]
ax.plot(total_scores)
ax.set_ylabel('Convex Hull Volume', fontsize=15)
ax.set_xlabel('Iterations of BayesOpt', fontsize=15)

ax=axs[1]
sns.scatterplot(data=data, x='x', y='y', hue='class', ax=ax, palette = 'Blues')
# for h in hulls:
#     ax.plot(h.points[h.vertices, 0], h.points[h.vertices, 1], 'r--', lw=2)
ax.set_ylabel(r'$y_{initial}$', fontsize=15)
ax.set_xlabel(r'$x_{initial}$', fontsize=15)

fig.tight_layout()
plt.show()
```
